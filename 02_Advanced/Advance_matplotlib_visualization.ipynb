###1. Colored Scatterplots

```python
iris = pd.read_csv('/content/iris.csv')
iris.info()
iris.tail()
```

```python
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(10)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('Sepal Lenght')
plt.ylabel('Patel Length')
```

<p>Visit this link for more information <a href='https://matplotlib.org/stable/users/explain/colors/colormaps.html'>Matplotlib cmap</a></p>

```python
#colored scatter
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')    #cmap='summer'/'viridis(By default)'
plt.xlabel('Sepal Lenght')
plt.ylabel('Patel Length')
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet')
plt.xlabel('Sepal Lenght')
plt.ylabel('Patel Length')
plt.colorbar()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=.2)   #alpha range between 0 to 1
plt.xlabel('Sepal Lenght')
plt.ylabel('Patel Length')
plt.colorbar()
```

###2. plot size


```python
plt.figure(figsize=(15,9))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=.7)
plt.xlabel('Sepal Lenght')
plt.ylabel('Patel Length')
plt.colorbar()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=1)
plt.xlabel('Sepal Lenght')
plt.ylabel('Patel Length')
plt.colorbar()
```

###8. 3D surface plots

<h3>Visit <a href='https://www.desmos.com/calculator/ezx5zyqip0'>Desmos graph</a></h3>

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure()
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='jet')
fig.colorbar(p)    #litte amount is different
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='winter')
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure()
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)    #litte amount is different
plt.show()
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure()
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)    #litte amount is different
plt.show()
```

###9. Contour plots

<p>#. convert 3D graph to 2D graph</p>

```python
fig = plt.figure()
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure()
ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

###10. Heatmap

```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
temp_df.sample(10)
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(10,10))
plt.imshow(grid)
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.colorbar()
```
