###Bar chart
<p>1. Bivariate Analysis</p>
<p>2. Numerical Vs Categorical</p>
<p>3. Use case -> Aggregate analysis of groups</p>

```python
#simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yello','pink']

plt.bar(colors,children,color='#BA55D3')
```

```python
#horizontal bar chart
plt.barh(colors,children,color='DarkMagenta')
```

```python
df = pd.read_csv('/content/batsman_season_record.csv')
df.head()
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(np.arange(df.shape[0]) - .2 , df['2015'],width=.2)
plt.bar(np.arange(df.shape[0])  , df['2016'],width=.2)
plt.bar(np.arange(df.shape[0]) + .2 , df['2017'],width=.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
#color and label
```

```python
#bar chart using data
```

```python
#A problem
children = [10,20,40,10,30]
colors = ['red red red red','blue blue blue blue blue','green green ','yello','pink']

plt.bar(colors,children,color='#BA55D3')
plt.xticks(rotation=90)  #rotation = 'vertical' is also allowed
```

```python
#stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],label='2015')
plt.legend()
plt.show()
```

###Pie chart
<p>1. Univariate/Bivariate Analysis</p>
<p>2. Categorical vs numerical</p>
<p>3. Use case -> To find contibution on a standard scale</p>

```python
#simple Example
data=[23,45,100,20,49]
subjects=['eng','math','bangla','eee','cse']
plt.pie(data,labels=subjects)
plt.show()
```

```python
gayle = pd.read_csv('/content/gayle-175.csv')
gayle
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct="%0.1f%%")
plt.show()
```

```python
#parcentage and colors
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct="%0.1f%%",colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
#Explode shadow
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct="%0.1f%%",explode=[0.1,0,0,0,0,0])
plt.show()
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct="%0.1f%%",explode=[0.1,0,0,0,0,0.1])
plt.show()
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct="%0.1f%%",explode=[0,0.5,0,0,0,0])
plt.show()
```

```python
#Shadow
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct="%0.1f%%",explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```
